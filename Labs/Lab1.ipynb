## C++

```python
%%writefile lab1_stack_cpp.cpp
#include <iostream>
using namespace std;

class Stack {
    int* arr;
    int top;
    int capacity;

public:
    Stack(int size) {
        capacity = size;
        arr = new int[capacity];
        top = -1;
    }

    void push(int value) {
        if (top == capacity - 1) {
            cout << "Stack Overflow" << endl;
            return;
        }
        arr[++top] = value;
        cout << "Pushed: " << value << endl;
    }

    void pop() {
        if (top == -1) {
            cout << "Stack Underflow" << endl;
            return;
        }
        cout << "Popped: " << arr[top--] << endl;
    }

    void peek() {
        if (top == -1) {
            cout << "Stack is empty" << endl;
        } else {
            cout << "Top element: " << arr[top] << endl;
        }
    }

    void traverse() {
        if (top == -1) {
            cout << "Stack is empty" << endl;
            return;
        }
        cout << "Stack elements: ";
        for (int i = top; i >= 0; i--)
            cout << arr[i] << " ";
        cout << endl;
    }

    void search(int value) {
        for (int i = top; i >= 0; i--) {
            if (arr[i] == value) {
                cout << "Found at position: " << (top - i) << endl;
                return;
            }
        }
        cout << "Element not found" << endl;
    }

    ~Stack() {
        delete[] arr;
    }
};

int main() {
    int size;
    cout << "Enter stack size: ";
    cin >> size;

    Stack s(size);
    s.push(10);
    s.push(20);
    s.push(30);
    s.peek();
    s.traverse();
    s.search(20);
    s.pop();
    s.traverse();

    return 0;
}

```

## Python

```python

# ‚úÖ Python Implementation of Stack using List (Array)

stack = []

def push(value):
    stack.append(value)

def pop():
    if not stack:
        print("Stack Underflow")
    else:
        print("Popped:", stack.pop())

def peek():
    if not stack:
        print("Stack is Empty")
    else:
        print("Top Element:", stack[-1])

def traverse():
    if not stack:
        print("Stack is Empty")
    else:
        print("Stack (Top to Bottom):", stack[::-1])

def search(value):
    if value in stack:
        pos = len(stack) - stack[::-1].index(value) - 1
        print(f"Found {value} at position {pos} from top")
    else:
        print(f"{value} not found in stack")

# üîç Test
push(10)
push(20)
push(30)
traverse()
peek()
search(20)
pop()
traverse()
```

```python

# Write the C program into a file
c_code = """
#include <stdio.h>
#define SIZE 100

int stack[SIZE];
int top = -1;

void push(int value) {
    if (top == SIZE - 1)
        printf("Stack Overflow\\n");
    else
        stack[++top] = value;
}

void pop() {
    if (top == -1)
        printf("Stack Underflow\\n");
    else
        printf("Popped: %d\\n", stack[top--]);
}

void peek() {
    if (top == -1)
        printf("Stack is Empty\\n");
    else
        printf("Top Element: %d\\n", stack[top]);
}

void traverse() {
    if (top == -1)
        printf("Stack is Empty\\n");
    else {
        printf("Stack: ");
        for (int i = top; i >= 0; i--)
            printf("%d ", stack[i]);
        printf("\\n");
    }
}

void search(int value) {
    for (int i = top; i >= 0; i--) {
        if (stack[i] == value) {
            printf("Found %d at position %d from top\\n", value, top - i);
            return;
        }
    }
    printf("%d not found in stack\\n", value);
}

int main() {
    push(10);
    push(20);
    push(30);
    traverse();
    peek();
    search(20);
    pop();
    traverse();
    return 0;
}
"""

with open("stack_lab1.c", "w") as f:
    f.write(c_code)
```
