 # **LAB- 5:- Queue Implementation with Operations**


**ðŸŽ¯Objective:**

* Implement queue and its operations like enqueue, dequeue, traverse, search

**Concept:**

A queue is a linear data structure that follows the FIFO (First In, First Out) principle.
In this example, we use an array-based circular queue

ðŸ§  Operations Covered:
* enqueue() â€“ Insert at rear

* dequeue() â€“ Remove from front

* traverse() â€“ Print all elements

* search() â€“ Find an element

 **C++ Code**

```python
%%writefile lab5_queue.cpp
#include <iostream>
#define SIZE 5
using namespace std;

class Queue {
    int arr[SIZE], front, rear;
public:
    Queue() {
        front = rear = -1;
    }

    void enqueue(int val) {
        if ((rear + 1) % SIZE == front) {
            cout << "Queue Overflow\n";
            return;
        }
        if (front == -1) front = 0;
        rear = (rear + 1) % SIZE;
        arr[rear] = val;
    }

    void dequeue() {
        if (front == -1) {
            cout << "Queue Underflow\n";
            return;
        }
        cout << "Dequeued: " << arr[front] << endl;
        if (front == rear)
            front = rear = -1;
        else
            front = (front + 1) % SIZE;
    }

    void traverse() {
        if (front == -1) {
            cout << "Queue is empty\n";
            return;
        }
        cout << "Queue: ";
        int i = front;
        while (true) {
            cout << arr[i] << " ";
            if (i == rear) break;
            i = (i + 1) % SIZE;
        }
        cout << endl;
    }

    void search(int key) {
        if (front == -1) {
            cout << "Queue is empty\n";
            return;
        }
        int i = front;
        while (true) {
            if (arr[i] == key) {
                cout << "Element " << key << " found at index " << i << endl;
                return;
            }
            if (i == rear) break;
            i = (i + 1) % SIZE;
        }
        cout << "Element not found\n";
    }
};

int main() {
    Queue q;
    q.enqueue(10);
    q.enqueue(20);
    q.enqueue(30);
    q.traverse();
    q.dequeue();
    q.traverse();
    q.search(20);
    q.search(99);
    return 0;
}
```

**C code**

```python
%%writefile lab5_queue.c
#include <stdio.h>
#define SIZE 5

int queue[SIZE];
int front = -1, rear = -1;

void enqueue(int val) {
    if ((rear + 1) % SIZE == front) {
        printf("Queue Overflow\n");
        return;
    }
    if (front == -1) front = 0;
    rear = (rear + 1) % SIZE;
    queue[rear] = val;
    printf("Enqueued: %d\n", val);
}

void dequeue() {
    if (front == -1) {
        printf("Queue Underflow\n");
        return;
    }
    printf("Dequeued: %d\n", queue[front]);
    if (front == rear)
        front = rear = -1;
    else
        front = (front + 1) % SIZE;
}

void traverse() {
    if (front == -1) {
        printf("Queue is empty\n");
        return;
    }
    printf("Queue: ");
    int i = front;
    while (1) {
        printf("%d ", queue[i]);
        if (i == rear) break;
        i = (i + 1) % SIZE;
    }
    printf("\n");
}

void search(int key) {
    if (front == -1) {
        printf("Queue is empty\n");
        return;
    }
    int i = front;
    while (1) {
        if (queue[i] == key) {
            printf("Element %d found at position %d\n", key, i);
            return;
        }
        if (i == rear) break;
        i = (i + 1) % SIZE;
    }
    printf("Element %d not found\n", key);
}

int main() {
    enqueue(10);
    enqueue(20);
    enqueue(30);
    traverse();
    dequeue();
    traverse();
    search(20);
    search(50);
    return 0;
}
```
